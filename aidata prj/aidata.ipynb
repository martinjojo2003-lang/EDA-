Q1
#What is the distribution of salaries (USD) across AI job roles?

```python
# df["job_title_clean"] = df["job_title"].astype(str).str.strip().str.title()

# Group salaries by job role and get summary statistics
salary_by_role = df.groupby("job_title")["salary_usd"].describe()
# Show first 10 rows of summary
print(salary_by_role.head(5))
# Plot distribution of all salaries
plt.figure(figsize=(8,5))
df["salary_usd"].plot(kind="hist", bins=50, edgecolor="black")
plt.title("Salary Distribution across AI Jobs")
plt.xlabel("Salary (USD)")
plt.ylabel("Number of Jobs")
plt.tight_layout()
plt.show()
```

```python
#Q2 Which job titles have the highest average salary?
# Group by job title and calculate mean salary
avg_salary_by_role = df.groupby("job_title")["salary_usd"].mean()
# Sort descending to get highest average salaries
top10_avg_salary_roles = avg_salary_by_role.sort_values(ascending=False).head(10)

print("Top 10 Job Titles by Average Salary (USD):")
print(top10_avg_salary_roles)
```

```python
plt.figure(figsize=(7,6))
top10_avg_salary_roles.plot(kind="bar", edgecolor="black")

plt.title("Top 10 Job Titles by Average Salary (USD)")
plt.xlabel("Job Title")
plt.ylabel("Average Salary (USD)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#Q3 How do salaries vary by experience level (Junior, Mid, Senior, Expert)?
# Group by experience level and calculate salary stats
salary_by_experience = df.groupby("experience_level")["salary_usd"].describe()
print(salary_by_experience)

# --- Plot: Bar chart of average salary ---
avg_salary_exp = df.groupby("experience_level")["salary_usd"].mean().sort_values()

plt.figure(figsize=(8,5))
sns.barplot(x=avg_salary_exp.index, y=avg_salary_exp.values)

plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary (USD)")
plt.tight_layout()
plt.show()

# --- Plot: Boxplot to show distribution ---
plt.figure(figsize=(8,5))
sns.boxplot(x="experience_level", y="salary_usd", data=df)

plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.tight_layout()
plt.show()
```

```python
#Q4 Does remote ratio (onsite vs remote) affect salary levels?
df["remote_category"] = df["remote_ratio"].map({0:"Onsite",  100:"Remote"})

# --- Single Graph: Boxplot ---
plt.figure(figsize=(5,5))
sns.boxplot(x="remote_category", y="salary_usd", data=df, palette="Set2")

plt.title("Salary Distribution by Work Mode (Onsite vs Remote)")
plt.xlabel("Work Mode")
plt.ylabel("Salary (USD)")
plt.tight_layout()
plt.show()
```

```python
#Q5 Which countries (company_location) hire the most AI professionals?
# Count job postings by company_location
country_counts = df['company_location'].value_counts().head(15)

# Plot top 15 countries hiring AI professionals
plt.figure(figsize=(5,4))
country_counts.plot(kind='bar', color='skyblue')
plt.title("Top 15 Countries Hiring AI Professionals")
plt.xlabel("Country")
plt.ylabel("Number of Job Postings")
plt.xticks(rotation=45)
plt.show()
```

```python
#Q6 Is there a relationship between education_required and salary_usd?
plt.figure(figsize=(5,4))

# Boxplot of salary by education level
sns.boxplot(x="education_required", y="salary_usd", data=df, palette="Set2", showfliers=False)

# Overlay average salaries (mean points)
edu_salary = df.groupby("education_required")["salary_usd"].mean()
sns.pointplot(x=edu_salary.index, y=edu_salary.values, color="red", markers="D", linestyles="")

plt.title("Relationship Between Education Level and Salary")
plt.xlabel("Education Level")
plt.ylabel("Salary (USD)")
plt.xticks(rotation=45)
plt.legend(["Average Salary"], loc="upper right")
plt.show()
```

```python
#Q7 How does years_experience impact salary?
plt.figure(figsize=(5,4))

# Scatter plot with regression line
sns.regplot(x="years_experience", y="salary_usd", data=df, scatter_kws={"alpha":0.4}, line_kws={"color":"red"})

plt.title("Impact of Years of Experience on Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (USD)")
plt.show()

# Grouped average salary by years of experience
avg_salary_exp = df.groupby("years_experience")["salary_usd"].mean()

plt.figure(figsize=(5,4))
avg_salary_exp.plot(kind="line", marker="o", color="blue")
plt.title("Average Salary by Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Average Salary (USD)")
plt.grid(True)
plt.show()

# Print summary
print(avg_salary_exp.head(2))
```

```python
#Q8 Which job titles have the longest job descriptions?
job_desc_length = df.groupby("job_title")["job_description_length"].mean().sort_values(ascending=False).head(5)

# Plot top 15 job titles with longest job descriptions
plt.figure(figsize=(5,4))
job_desc_length.plot(kind="bar", color="purple")
plt.title("Top 15 Job Titles with Longest Job Descriptions")
plt.xlabel("Job Title")
plt.ylabel("Average Job Description Length (characters)")
plt.xticks(rotation=75)
plt.show()

# Print results
print(job_desc_length)
```

```python
#Q9 Do companies with higher benefits_score also offer higher
# Drop rows with missing values in the relevant columns
df_bs = df.dropna(subset=["benefits_score", "salary_usd"])

# Scatter plot with smoothing (regression line)
plt.figure(figsize=(5,4))
sns.regplot(x="benefits_score", y="salary_usd", data=df_bs, scatter_kws={"alpha":0.4}, line_kws={"color":"red"})
plt.title("Salary (USD) vs Benefits Score")
plt.xlabel("Benefits Score")
plt.ylabel("Salary (USD)")
plt.grid(True)
plt.show()

# Compute correlations
pearson = df_bs["benefits_score"].corr(df_bs["salary_usd"], method="pearson")
spearman = df_bs["benefits_score"].corr(df_bs["salary_usd"], method="spearman")

print(f"Pearson correlation (benefits_score, salary_usd): {pearson:.4f}")
print(f"Spearman correlation (benefits_score, salary_usd): {spearman:.4f}")

# Group by rounded benefits score to see average salaries
df_bs["benefits_rounded"] = df_bs["benefits_score"].round(0)
grouped = df_bs.groupby("benefits_rounded")["salary_usd"].agg(["count", "mean", "median"]).sort_index()
print("\nAverage salary by rounded benefits_score:\n", grouped)
```

```python
#Q10 How does employment_type (CT, PT, FL, etc.) affect salaries?
# Keep relevant columns and drop NA
df_et = df.dropna(subset=["employment_type", "salary_usd"])

# Order employment types by median salary for clearer plotting
order = df_et.groupby("employment_type")["salary_usd"].median().sort_values(ascending=False).index

plt.figure(figsize=(5,4))
sns.boxplot(x="employment_type", y="salary_usd", data=df_et, order=order, showfliers=False)
sns.stripplot(x="employment_type", y="salary_usd", data=df_et, order=order, color="black", size=3, alpha=0.3, jitter=True)
plt.title("Salary distribution by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Salary (USD)")
plt.xticks(rotation=45)
plt.show()

# Print summary statistics
summary = df_et.groupby("employment_type")["salary_usd"].agg(["count", "mean", "median", "std"]).loc[order]
print("Salary summary by employment_type:\n", summary)
```

```python
#Q11 What are the top companies hiring the most AI professionals?
# Count job postings by company_name
top_companies = df["company_name"].value_counts().head(5)  # top 20
print("Top 20 companies by number of postings:\n", top_companies)

# If you want a bar plot:
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
top_companies.plot(kind="bar")
plt.title("Top 20 Companies Hiring the Most AI Professionals")
plt.xlabel("Company")
plt.ylabel("Number of Postings")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
#Q12 How do application deadlines vary across industries?
# Keep rows where both dates are valid
# Parse dates safely
df_dates = df.copy()
df_dates["posting_date"] = pd.to_datetime(df_dates["posting_date"], errors="coerce")
df_dates["application_deadline"] = pd.to_datetime(df_dates["application_deadline"], errors="coerce")

# Keep rows where both dates are valid
df_dates = df_dates.dropna(subset=["posting_date", "application_deadline", "industry"])

# Compute days until deadline
df_dates["days_until_deadline"] = (df_dates["application_deadline"] - df_dates["posting_date"]).dt.days

# Remove obviously negative or extreme values optionally
df_dates = df_dates[(df_dates["days_until_deadline"] >= 0) & (df_dates["days_until_deadline"] <= 365)]

# Average and median days until deadline by industry
industry_stats = df_dates.groupby("industry")["days_until_deadline"].agg(["count", "mean", "median", "std"]).sort_values("mean", ascending=False)
print("Application deadline stats by industry (count, mean days, median days, std):\n", industry_stats)

# Plot top industries by count (to focus on meaningful stats)
top_industries = industry_stats.sort_values("count", ascending=False).head(6).index
plt.figure(figsize=(8,6))
sns.boxplot(x="industry", y="days_until_deadline", data=df_dates[df_dates["industry"].isin(top_industries)], order=top_industries, showfliers=False)
plt.title("Distribution of Days Until Application Deadline by Industry (Top 12 Industries by Count)")
plt.xlabel("Industry")
plt.ylabel("Days until deadline (application_deadline - posting_date)")
plt.xticks(rotation=75)
plt.show()
```

Conclusion: Key Insights and Future Directions

The Exploratory Data Analysis of the AI job postings dataset provided a rich and detailed view of the modern AI workforce. Our analysis successfully identified several key trends and defining characteristics of the industry.

Summary of Key Findings
Compensation Drivers: The data clearly demonstrated a strong positive correlation between experience level (SE/EX) and salary_usd, with advanced degrees (PhD/Master) also correlating with higher pay, especially in research-intensive roles.

Skill Demand: Analysis of the required_skills column highlighted the perennial demand for core technologies such as Python, SQL, and Deep Learning, while also showing an emerging need for skills in specialized areas like MLOps and cloud platforms like AWS and Azure.

Geographical Concentration: A significant concentration of high-paying jobs was observed in major tech hubs (as indicated by company_location), though the remote_ratio and employee_residence fields pointed to a growing flexibility in where AI professionals can be based.

Implications and Future Work
